```python
# Dataset from Kaggle, created by Leonardo Henrique
# Load libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

tracks = pd.read_csv('/Users/joerussell/Desktop/DATASCIENCE/DATASETS/spotify-tracks.csv')
artists = pd.read_csv('/Users/joerussell/Desktop/DATASCIENCE/DATASETS/spotify-artists.csv')
```

```python
tracks.sort_values(by='popularity', ascending=False)
```

```python
# Most popular songs are the most recent ones, presumably because popularity is a measure of current popularity,
# so drop popularity features

tracks = tracks.drop('popularity', axis=1)
```

```python
# Retrieve years from release_date feature

years = tracks['release_date'].str.split('-')
tracks['year'] = [i[0] for i in years]
tracks.year = tracks.year.astype(int)
```

```python
tracks.year.value_counts()
```

```python
# Drop outlier
tracks = tracks.drop(labels=tracks[tracks.year == 1900].index, axis=0)
```

```python
# Correlations

tracks.corr()
```

```python
corr_matrix = tracks.corr()
sns.heatmap(corr_matrix)

plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Lineplot of highly correlated features

# Scale loudness, since its range is much different from the other variables
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
tracks[['loudness']] = scaler.fit_transform(tracks[['loudness']])

highcorr_features = ['explicit', 'danceability', 'energy', 'acousticness', 'instrumentalness', 'loudness']
x = [i for i in range(1900, 2022)]
for feature in highcorr_features:
    y = [np.mean(tracks[feature][tracks.year == i]) for i in x]
    sns.lineplot(x=x, y=y, label=feature)

plt.title('Feature Averages Over Time')
plt.show()
```

```python
# Plot mean energy by year vs year

x = [i for i in range(1900, 2022)]
y = [np.mean(tracks.energy[tracks.year == i]) for i in x]
sns.regplot(x=x, y=y, order=3)

plt.title('Song Energy Over Time')
plt.xlabel('Year')
plt.ylabel('Energy')
plt.show()
```

```python
# Starts a bit scattered, but by 1960, this follows a clear trend. The correlation was high (0.46), but this plot
# indicates this non-linear relationship is even stronger.
```

```python
# Plot mean acousticness by year vs year

x = [i for i in range(1900, 2022)]
y = [np.mean(tracks.acousticness[tracks.year == i]) for i in x]
sns.regplot(x=x, y=y, order=3)

plt.title('Song Acousticness Over Time')
plt.xlabel('Year')
plt.ylabel('Acousticness')
plt.show()
```

```python
# Another strong non-linear relationship. These two plots are maybe more scattered in the earlier years because
# maybe there are fewer songs in the dataset for those years?
```
