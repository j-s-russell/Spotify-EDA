```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

tracks = pd.read_csv('/Users/joerussell/Desktop/DATASCIENCE/DATASETS/spotify-tracks.csv')
```

```python
tracks.head()
```

```python
tracks.isnull().sum()
```

Song names aren't relevant for this analysis, so missing values aren't an issue.

```python
tracks.sort_values(by='popularity', ascending=False)
```

'popularity' presumably measures current popularity, so we can drop this feature.

```python
tracks = tracks.drop('popularity', axis=1)
```

```python
# Retrieve years from release_date feature

years = tracks['release_date'].str.split('-')
tracks['year'] = [i[0] for i in years]
tracks.year = tracks.year.astype(int)
```

```python
tracks.year.value_counts()
```

```python
# Drop outlier
tracks = tracks.drop(labels=tracks[tracks.year == 1900].index, axis=0)
```

```python
# Correlations

tracks.corr()
```

```python
corr_matrix = tracks.corr()
sns.heatmap(corr_matrix)

plt.title('Feature Correlation Matrix')
plt.show()
```

```python
# Lineplot of highly correlated features

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
tracks[['loudness']] = scaler.fit_transform(tracks[['loudness']])

highcorr_features = ['explicit', 'danceability', 'energy', 'acousticness', 'instrumentalness']
x = [i for i in range(1900, 2022)]
for feature in highcorr_features:
    y = [np.mean(tracks[feature][tracks.year == i]) for i in x]
    sns.lineplot(x=x, y=y, label=feature)

plt.title('Feature Averages Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# Plot mean energy by year vs year

x = [i for i in range(1900, 2022)]
y = [np.mean(tracks.energy[tracks.year == i]) for i in x]
sns.regplot(x=x, y=y, order=3)

plt.title('Song Energy Over Time')
plt.xlabel('Year')
plt.ylabel('Energy')
plt.show()
```

The plot starts a bit scattered, but by 1960, this follows a clear trend. The correlation was high (0.46), and this plot indicates this non-linear relationship is strong.

```python
# Plot mean acousticness by year vs year

x = [i for i in range(1900, 2022)]
y = [np.mean(tracks.acousticness[tracks.year == i]) for i in x]
sns.regplot(x=x, y=y, order=3)

plt.title('Song Acousticness Over Time')
plt.xlabel('Year')
plt.ylabel('Acousticness')
plt.show()
```

Acousticness shows another strong relationship. Since the dataset includes fewer songs during earlier years, there is more variation in the above plots for songs from the early 1900's.

```python
# Plot mean loudness by year vs year

x = [i for i in range(1900, 2022)]
y = [np.mean(tracks.loudness[tracks.year == i]) for i in x]
sns.regplot(x=x, y=y, order=6)

plt.title('Song Loudness Over Time')
plt.xlabel('Year')
plt.ylabel('Loudness')
plt.show()
```

Starting in the mid-1900's, popular songs are shown to increase significantly in loudness and energy, while acousticness has decreased significantly during this time period as well.
